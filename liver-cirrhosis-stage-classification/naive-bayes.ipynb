## Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

## Import Dataset

```python
train_df = pd.read_csv('data_train.csv')
x_train = train_df.drop(columns=['Stage'])
y_train = train_df['Stage']
x_train.head()
```

```python
test_df = pd.read_csv('data_test.csv')
x_test = test_df.drop(columns=['Stage'])
y_test = test_df['Stage']
x_test.head()
```

## Naive Bayess

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
```

```python
nb_pred = nb_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print('Naive Bayess Accuracy Score: ', accuracy_score(y_test, nb_pred))
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'var_smoothing': np.logspace(0,-9, num=100)
}

nb_grid = GridSearchCV(estimator=nb_model, param_grid=param_grid, cv=5)
nb_grid.fit(x_train, y_train)
```

```python
nb_grid.best_params_
```

```python
nb_model_grid = GaussianNB(var_smoothing=1.0)
nb_model_grid.fit(x_train, y_train)
```

```python
nb_model_grid_pred = nb_model_grid.predict(x_test)
```

```python
print('Naive Bayess Accuracy Score after tuning: ', accuracy_score(y_test, nb_model_grid_pred))
```
